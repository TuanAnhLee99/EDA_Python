##### 4.	Phương thức thanh toán (xét cột Payment) có ảnh hưởng đến số lượng đơn hàng hay không, vẽ biểu đồ minh họa và nhận xét kết quả

##### Ta nhận thấy rằng biến Payment là biến phân loại (Categorical variable) và Quantity là biến numerical variable, cho nên ta sẽ dùng kiểm định phân loại  ANOVA

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Tạo DataFrame mới với các cột Payment và Quantity
data_PQ = df[['Payment', 'Quantity']]

# Thực hiện phân tích ANOVA
model = ols('Quantity ~ C(Payment)', data=data_PQ).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# In bảng ANOVA
print(anova_table)
```

#### Kết quả phân tích ANOVA cho thấy:
Dựa vào giá trị p-value (0.973531) từ kết quả cho thấy giá trị p-value =0.973531 >> 0.05, nghĩa là không có đủ bằng chứng để bác bỏ giả thuyết không có ảnh hưởng của phương thức thanh toán đến số lượng đơn hàng. Do đó, không có ảnh hưởng đáng kể giữa phương thức thanh toán và số lượng đơn hàng.

Tóm lại, kết luận từ kết quả phân tích ANOVA là phương thức thanh toán Payment KHÔNG ảnh hưởng đến số lượng đơn hàng Quantity

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Tạo DataFrame mới với các cột Payment và Quantity
data_PQ = df[['Payment', 'Quantity']]

# Tính trung bình Quantity theo từng nhóm Payment
mean_quantity = data_PQ.groupby('Payment')['Quantity'].mean().reset_index()

# Tính độ lệch chuẩn và số lượng quan sát
std_quantity = data_PQ.groupby('Payment')['Quantity'].std().reset_index()
count_quantity = data_PQ['Payment'].value_counts().reset_index()

# Thiết lập màu sắc cho từng nhóm
colors = ['#4CAF50', '#9C27B0', '#2196F3']

# Vẽ biểu đồ cột với số liệu thống kê và màu sắc khác nhau
plt.figure(figsize=(8, 6))
sns.barplot(x='Payment', y='Quantity', data=mean_quantity, palette=colors, 
            yerr=std_quantity['Quantity'], capsize=0.1, alpha=0.8, 
            edgecolor='black', linewidth=1.5)
plt.title('Average Quantity by Payment')
plt.xlabel('Payment')
plt.ylabel('Average Quantity')

# Thêm tiêu đề cho trục x và trục y
plt.xticks(rotation=45)
plt.gca().yaxis.grid(True)
plt.gca().set_axisbelow(True)
plt.gca().set_ylim([0, max(mean_quantity['Quantity']) * 1.2])
plt.text(-0.2, max(mean_quantity['Quantity']) * 1.1, "n = {}".format(count_quantity['Payment'][0]), fontsize=10)
plt.text(0.8, max(mean_quantity['Quantity']) * 1.1, "n = {}".format(count_quantity['Payment'][1]), fontsize=10)
plt.text(1.8, max(mean_quantity['Quantity']) * 1.1, "n = {}".format(count_quantity['Payment'][2]), fontsize=10)

plt.show()
```

```python
# In giá trị trung bình của Quantity
for i in range(len(mean_quantity)):
    print("Average Quantity for {}: {:.2f}".format(mean_quantity['Payment'][i], mean_quantity['Quantity'][i]))
```

### 6.	Khung giờ nào là giờ cao điểm mua sắm trong ngày, vẽ đồ thị minh họa và nhận xét kết quả.

```python
import pandas as pd
import matplotlib.pyplot as plt



# Chuyển cột 'Time' sang định dạng thời gian
df['Time'] = pd.to_datetime(df['Time'])

# Trích xuất giờ từ cột 'Time'
df['Hour'] = df['Time'].dt.hour

# Đếm số lượng mua hàng theo từng khung giờ
hour_counts = df['Hour'].value_counts().sort_index()

# Vẽ biểu đồ minh họa
plt.figure(figsize=(10, 6))
plt.plot(hour_counts.index, hour_counts.values, marker='o', linestyle='-', color='blue', linewidth=2)

# Thêm tiêu đề và nhãn cho trục x và trục y
plt.title('Peak Shopping Hours', fontsize=16)
plt.xlabel('Hour', fontsize=12)
plt.ylabel('Quantity', fontsize=12)

plt.xticks(range(24), fontsize=10)  # Đặt các đỉnh trục x từ 0 đến 23

# Tăng độ dày của trục x và trục y
plt.gca().spines['bottom'].set_linewidth(1.5)
plt.gca().spines['left'].set_linewidth(1.5)

# Tăng độ dày của các đường viền trục x và trục y
plt.gca().tick_params(axis='x', width=1.5)
plt.gca().tick_params(axis='y', width=1.5)

# Tạo lưới trong biểu đồ
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

plt.show()
```
